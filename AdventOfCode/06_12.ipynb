```python
def get_sum():
    answers = open("source06_12.txt").read().splitlines()
    result = 0
    my_set = set()
    for answer in answers:
        if answer == '':
            result += len(my_set)
            my_set.clear()
        else : 
            my_set.update(list(answer))
    result += len(my_set)
    return result

get_sum()
```

```python
import string

def task2():
    answers = open("source06_12.txt").read().splitlines()
    result = 0
    my_set = set(string.ascii_lowercase)
    for answer in answers:
        if answer == '':
            result += len(my_set)
            my_set = set(string.ascii_lowercase)
        else : 
            my_set.intersection_update(list(answer))
    result += len(my_set) 
    return result
    
task2()
```
