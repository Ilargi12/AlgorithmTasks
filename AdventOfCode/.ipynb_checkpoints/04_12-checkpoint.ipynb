```python
import re

def check_year(number, start, end):
    return len(number) == 4 and start <= int(number) <= end

def check_height(height):
    pattern = re.compile("([0-9]+)([a-z]+)") 
    res = pattern.match(height)
    if res is None : return False
    res = res.groups()
    return (res[1] == 'cm' and 150 <= int(res[0]) <= 193) or (res[1] == 'in' and 59 <= int(res[0]) <= 76)
    
def check_hcl(hcl):
    pattern = re.compile("[a-z0-9]+")
    return len(hcl) == 7 and hcl[0] == '#' and pattern.fullmatch(hcl) is not None

def check_pid(pid):
    pattern = re.compile("[0]+[0-9]+")
    return len(pid) == 9 and pattern.fullmatch(pid) is not None

def check_dic(dic):
    keys = list(dic.keys())
    if len(keys) == 8 or (len(keys) == 7 and not 'cid' in keys):
        if (check_year(dic['byr'], 1920, 2020) and
        check_year(dic['iyr'], 2010, 2020) and
        check_year(dic['eyr'], 2020, 2030) and
        check_height(dic['hgt']) and 
        check_hcl(dic['hcl']) and 
        dic['ecl'] in ['amb', 'blu', 'brn', 'gey', 'grn', 'hzl', 'oth'] and
        check_pid(dic['pid'])):
            return True
    return False

def count_valid_passports():
    passports = open("source04_12.txt").read().splitlines()
    dic = {}
    counter = 0
    for line in passports:
        if line == '':
            print(dic)
            if check_dic(dic): counter += 1
            dic = {}
        else:
            array = line.split(" ")
            for field in array:
                f, s = field.split(":")
                dic[f] = s
    return counter
    
count_valid_passports()
```
